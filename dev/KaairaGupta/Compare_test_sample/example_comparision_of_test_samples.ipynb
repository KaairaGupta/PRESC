```python
import compare_model_classification as cmc
import numpy as np
```

```python
model_1 = []
model_2 = []

a = np.random.rand(12)
b = np.random.rand(12)

# adding arrays in model_1 and model_2
for i in range(len(a)):
    temp_1 = [a[i],1-a[i],1-int(round(a[i]))]
    if round(a[i]) == 0:
        temp_2 = [min(b[i],1-b[i]),max(b[i],1-b[i]),1]
    else:
        temp_2 = [max(b[i],1-b[i]),min(b[i],1-b[i]),0]
    
    model_1.append(temp_1)
    model_2.append(temp_2)
    
cmc.compare_model_classification(model_1,model_2)
```
