Example showing usage of the function `visualise_evaluation_metric` in `visualise_evaluation_metric.py`:

```python
import numpy as np
```

Format of input data:
[[x1,x2,x3,...,xn],
 [y1,y2,y3,...,yn],
 .
 .
 .
 [y1,y2,y3,...,yn]
]

Data must be numpy array.

```python
data = []

x = []

# Creating simple data for [x1,x2,x3,...x4]
for i in range(10):
    x.append(i)

# add x as data[0]
data.append(x)

# adding arrays for [y1,y2,y3,...yn]
for _ in range(len(x)):
    a = np.random.exponential(scale=1.0, size=10)
    data.append(a)

data = np.array(data)
```

```python
import visualise_evaluation_metric as vem

# Plotting evaluation_metric data
vem.visualise_evaluation_metric(data)
```
